# TASK B

Finding the common attributes

```python
#imports
import pandas as pd
from collections import Counter
import re
from itertools import combinations

file_path = 'beer_products.csv'
beer_data = pd.read_csv('/beer_products.csv')

# Preprocess the 'Text' column by removing non-alphabetical characters and converting text to lowercase
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    text = text.lower()  # Convert to lowercase
    return text

# Apply the preprocessing function to the 'Text' column
beer_data['Cleaned_Text'] = beer_data['Text'].apply(preprocess_text)

# Tokenize the cleaned text into individual words
all_words = ' '.join(beer_data['Cleaned_Text']).split()

# Get the frequency of each word
word_freq = Counter(all_words)

# Define stop words to filter out irrelevant words
stopwords = {'and', 'a', 'the', 'of', 'with', 'is', 'to', 'i', 'in', 'this'}

# Remove stop words from the word frequency count
filtered_word_freq = {word: freq for word, freq in word_freq.items() if word not in stopwords}

# Get the top 10 most frequent words related to beer attributes
top_words = [word for word, freq in Counter(filtered_word_freq).most_common(10)]
print("top words", top_words)
```

```python
# Function to clean text data and return tokenized words
def clean_and_tokenize(text):
    # Convert to lowercase, remove non-alphabetic characters, and split the text into words
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    words = text.split()
    return words

# Tokenizing all the reviews in the dataset
all_reviews = beer_data['Text'].dropna().apply(clean_and_tokenize)
# Ask the user to input 3 attributes for the analysis
user_attributes = input("Please enter three beer attributes separated by commas: ").split(',')
descriptive_words = [word.strip().lower() for word in user_attributes]

# Save the user input attributes to a file
with open('attributes.txt', 'w') as f:
    f.write(','.join(descriptive_words))

# Create a function to count co-occurrences of words in a review
def count_co_occurrences(review, word_pairs):
    return {pair: int(all(word in review for word in pair)) for pair in word_pairs}

# Create all possible combinations of the selected words for the lift analysis
word_pairs = list(combinations(descriptive_words, 2))

# Initialize counters for the word pairs and the total occurrences of individual words
pair_counts = {pair: 0 for pair in word_pairs}
word_totals = {word: 0 for word in descriptive_words}
total_reviews = len(all_reviews)

# Count occurrences of individual words and co-occurrences of word pairs
for review in all_reviews:
    # Convert review to a set of words for fast lookup
    review_words = set(review)
    # Update counts for individual words
    for word in descriptive_words:
        word_totals[word] += int(word in review_words)
    # Update counts for word pairs
    co_occurrences = count_co_occurrences(review_words, word_pairs)
    for pair, count in co_occurrences.items():
        pair_counts[pair] += count

# Calculate the lift for each word pair
lift_values = {}
for pair in word_pairs:
    word1, word2 = pair
    # Calculate the probability of each word individually and the pair together
    p_word1 = word_totals[word1] / total_reviews
    p_word2 = word_totals[word2] / total_reviews
    p_pair = pair_counts[pair] / total_reviews
    # Calculate lift (ratio of joint probability to the product of individual probabilities)
    lift_values[pair] = p_pair / (p_word1 * p_word2) if (p_word1 * p_word2) > 0 else 0

# Display the lift values for the word pairs
lift_values_sorted = sorted(lift_values.items(), key=lambda x: x[1], reverse=True)

# Print the three most unique beer attributes based on lowest lift values
print("\nThree most unique beer attribute pairs based on lift values:")
for pair, lift_value in lift_values_sorted[:3]:
    print(f"Attributes: {pair[0]} and {pair[1]} - Lift: {lift_value}")
```

# TASK C

Similarity Analysis

```python
#imports
from google.colab import files
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Read the attributes from the file
with open('attributes.txt', 'r') as f:
    customer_attributes = f.read().split(',')

# Combine the reviews and customer attributes into a single corpus
corpus = beer_data['Cleaned_Text'].tolist() + [' '.join(customer_attributes)]

# Create a bag-of-words model using CountVectorizer
vectorizer = CountVectorizer().fit(corpus)

# Transform the reviews and customer attributes into vectors
review_vectors = vectorizer.transform(beer_data['Cleaned_Text'])
attribute_vector = vectorizer.transform([' '.join(customer_attributes)])

# Calculate cosine similarity between each review and the customer-specified attributes
cosine_similarities = cosine_similarity(review_vectors, attribute_vector)

# Add the similarity scores to the DataFrame
beer_data['Similarity_Score'] = cosine_similarities

# Create a new DataFrame with product_name, review, and similarity_score columns
output_df = beer_data[['Beer Product', 'Text', 'Similarity_Score']]
output_df.columns = ['product_name', 'product_review', 'similarity_score']

# Sort the DataFrame by similarity_score in descending order
sorted_df = output_df.sort_values(by='similarity_score', ascending=False)

# Save the sorted results to a new CSV file
sorted_df.to_csv('sorted_beer_similarity_scores.csv', index=False)

# Print the top 5 rows of the sorted DataFrame to verify
sorted_df.head()
```

```python
# Load the dataset
file_path = 'beer_products.csv'
beer_data = pd.read_csv(file_path)

# Function to clean text data
def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text.lower())

# Check for distinct beer products in the dataset
unique_beers = beer_data['Beer Product'].unique()

# Select the first 10 unique beer products
ten_beers = unique_beers[:10]

# Filter the dataset to include only the 10 selected beers
selected_beers_data = beer_data[beer_data['Beer Product'].isin(ten_beers)]

# Clean the reviews
selected_beers_data['cleaned_text'] = selected_beers_data['Text'].apply(lambda x: clean_text(str(x)))

# Group the reviews by beer product to combine all reviews for each beer
grouped_reviews = selected_beers_data.groupby('Beer Product')['cleaned_text'].apply(' '.join).reset_index()

# Select the first beer for comparison: "Toppling Goliath Kentucky Brunch"
beer_to_compare = grouped_reviews[grouped_reviews['Beer Product'] == "Toppling Goliath Kentucky Brunch"]

# Select the remaining beers to compare against
remaining_beers = grouped_reviews[grouped_reviews['Beer Product'] != "Toppling Goliath Kentucky Brunch"]

# Vectorize the reviews using Bag-of-Words model
vectorizer = CountVectorizer()
vectorized_reviews = vectorizer.fit_transform(grouped_reviews['cleaned_text'])

# Calculate cosine similarity for "Toppling Goliath Kentucky Brunch" against the rest
similarity_matrix = cosine_similarity(vectorized_reviews)

# Extract the similarity scores for "Toppling Goliath Kentucky Brunch" (first beer)
toppling_index = grouped_reviews.index[grouped_reviews['Beer Product'] == "Toppling Goliath Kentucky Brunch"][0]
similarity_scores = similarity_matrix[toppling_index]

# Exclude the self-similarity (similarity with itself)
similarity_scores[toppling_index] = -1  # Set its own similarity to a low value to exclude it

# Find the beer with the highest similarity (excluding itself)
most_similar_beer_index = similarity_scores.argmax()  # This will give the index of the most similar beer
most_similar_beer = grouped_reviews.iloc[most_similar_beer_index]['Beer Product']

# Print the most similar beer
print(f"The most similar beer to 'Toppling Goliath Kentucky Brunch' is: {most_similar_beer}")
```
